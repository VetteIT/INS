# Úloha č.1

### Znenie

Definujte a vytvorte sieť, ktorá bude akceptovať vstupné dáta vo forme definovaného tensora a obsahuje lineárnu výstupnú vrstvu o veľkosti 5. Veľkosť vstupnej vrstvy zistite na základe vstupných dát, nezabúdajte však na to, že **nn.Linear** akceptuje iba jednorozmerné dáta, preto (v implementácii dopredného prechodu), vstupný tensor správne pretvarujte (podľa cvičenia 1). Pozorujte, aké výstupné hodnoty sieť vracia na základe definovaného tensora.

### Riešenie


```python
import torch
from torch import nn

input_data = torch.rand(4,4)

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.linear = nn.Linear(16, 5) 

    def forward(self, x):
        x = torch.flatten(x)
        x = self.linear(x)
        x = torch.relu(x)
        
        return x
    
model = Net()
y = model(input_data)
y
```

### Vysvetlenie

Sieť je vytvorená rozšírením triedy *nn.Module*. Obsahuje len jednu lineárnu vrstvu, s vstupnou veľkosťou 16 (pretože dáta majú formu tensorov o veľkosti 4\*4) a výstupnou veľkosťou 5, podľa zadania. Dáta sú v doprednom prechode najprv transformované na 1D tensor o veľkosti 16, pomocou funkcie *torch.flatten()*. Takéto dáta prechádzajú cez lineárnu vrstvu a následne aktivačnú funkciu ReLU. Výsledkom dopredného prechodu je teda tensor o dĺžke 5, obsahujúci nezáporné logity.

# Úloha č.2

### Znenie

Sieť ktorú ste vytvorili v úlohe č.1 rozšírte o dve skryté vrstvy, o veľkostiach **20** a **10**

### Riešenie

```python
import torch
from torch import nn

input_data = torch.rand(4,4)

class NewNet(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.hidden1 = nn.Linear(16, 20) 
        self.hidden2 = nn.Linear(20, 10) 
        self.output = nn.Linear(10, 5) 

    def forward(self, x):
        x = torch.flatten(x)
        
        x = self.hidden1(x)
        x = torch.relu(x)
        
        x = self.hidden2(x)
        x = torch.relu(x)
        
        x = self.output(x)
        x = torch.relu(x)
        
        return x
    
model = Net()
y = model(input_data)
y
```

### Vysvetlenie

Do siete boli pridané dve lineárne vrstvy. Dáta v poprednom prechode najprv prechádzajú cez vrstvu *hidden1*, následne cez *hidden2* a nakoniec cez *output*. Aby sieť mohla fungovať, je vstupná veľkost vždy nastavená na výstupnú veľkosť predchádzajúcej vrstvy.

# Úloha č.3

### Znenie

Refaktorujte vami vytvorenú sieť z úlohy č.2 tak, aby zodpovedala hore ukázanému prístupu vytvárania sietí. Do kódu dplňte parametre siete, vytvorenie jej vrstiev a dopredný prechod.

### Riešenie

```python
import torch
from torch import nn

input_data = torch.rand(4,4)

class NewNet(nn.Module):
    def __init__(self, input_size, hidden_size1, hidden_size2, output_size):
        super(NewNet, self).__init__()
        self.hidden1 = nn.Linear(input_size, hidden_size1) 
        self.hidden2 = nn.Linear(hidden_size1, hidden_size2) 
        self.output = nn.Linear(hidden_size2, output_size) 

    def forward(self, x):
        x = torch.flatten(x)
        
        x = self.hidden1(x)
        x = torch.relu(x)
        
        x = self.hidden2(x)
        x = torch.relu(x)
        
        x = self.output(x)
        x = torch.relu(x)
        
        return x
    
model = NewNet(16, 20, 10, 5)
y = model(input_data)
y
```

### Vysvetlenie

Sieť bola refaktorovaná tak, aby už neobsahovala tzv. "hardcoded" parametre jej vrstiev. Tie sú po refaktorizácií zadávané ako parametre konštruktora siete. Sieť je vďaka tomu jednoduchšie upravovateľná.

# Úloha č.4

Vytvorte novú sieť s rovnakou štruktúrou vrstiev ako sieť, ktorú ste vytvorili v **úlohe č.3**. V novej sieti ale použite aktivačnú fukncie **tanh**. Nakoniec vykonajte pomocou oboch sietí perdikciu na rovnakých náhodných dátach a porovnajte, či to ovplyvnilo výsledok predikcie.

```python
class NewNetTanh(nn.Module):
    def __init__(self, input_size, hidden_size1, hidden_size2, output_size):
        super(NewNetTanh, self).__init__()
        self.hidden1 = nn.Linear(input_size, hidden_size1) 
        self.hidden2 = nn.Linear(hidden_size1, hidden_size2) 
        self.output = nn.Linear(hidden_size2, output_size) 

    def forward(self, x):
        x = torch.flatten(x)
        
        x = self.hidden1(x)
        x = torch.tanh(x)
        
        x = self.hidden2(x)
        x = torch.tanh(x)
        
        x = self.output(x)
        x = torch.tanh(x)
        
        return x
    
model1 = NewNet(16, 20, 10, 5)
model2 = NewNetTanh(16, 20, 10, 5)

input_data = torch.rand(4,4)

y1 = model1(input_data)
y2 = model2(input_data)

print(y1)
print(y2)
```

### Vysvetlenie

Nová sieť bola vytvorená na základe siete *NewNet* z predchádzajúcej úlohy, avšak za použitia aktivačnej funkcie *tanh*. To sa na výstupe siete odzrkadlilo tak, že výstupom novej siete *NewNetTanh* už nie je tensor nezáporných logitov, ale tensor logitov v rozsahu -1 až 1. Vyplýva to z oboru hodnôt použitej aktivačnej funkcie *tanh*.
